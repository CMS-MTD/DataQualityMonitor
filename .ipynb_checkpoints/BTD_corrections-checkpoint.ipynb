```python
import numpy as np
import os, re, shutil
import ROOT as rt

from sklearn.ensemble import GradientBoostingRegressor

from root_numpy import root2array
from lib.histo_utilities import create_TH1D, create_TH2D
from lib.cebefo_style import cebefo_style
```

```python
cebefo_style()
```

```python
input_file = '../data/BTL_March_Testbeam/DataRun145.root'
ch = 14
```

```python
data = root2array(input_file)
```

```python
canvas = {}
canvas['amp'] = {}
canvas['int'] = {}
canvas['wave'] = {}
canvas['pos'] = {}
canvas['w_pos'] = {}

info = {ch: {}, 9:{}}
```

```python
for k in [9, ch]:
    
    '''=========================== Amplitude ==========================='''
    canvas['amp'][k] = rt.TCanvas('c_amp_'+str(k), 'c_amp_'+str(k), 800, 600)
    
    name = 'h_amp_'+str(k)
    title = 'Amplitude channel '+str(k)
    amp = (data['amp'].T)[k]
    h = create_TH1D(amp, name, title, 
                    binning = [100, 0, 550],
                    axis_title = ['Peak amplitude [mV]', 'Events / 5.5 mV'])
            
    h.GetXaxis().SetRange(int(40/5.5)+1,int(450/5.5)+1)
    i_max = h.GetMaximumBin()
    h.GetXaxis().SetRange(1,100)
    peak = h.GetBinCenter(i_max)
    info[k]['amp_range'] = [max(20,peak*0.6), min(450, peak*1.6)]
    res = h.Fit('landau','LQSR', '', info[k]['amp_range'][0], info[k]['amp_range'][1])
    
    if(h.GetMaximum() - h.GetMinimum() > 1000):
        canvas['amp'][k].SetLogy()
    h.DrawCopy('E1')
    canvas['amp'][k].Update()    
    canvas['amp'][k].Draw()
```

```python
selection = 'amp[9]>50 && amp[9]<200 && amp['+str(ch)+']>170 && amp['+str(ch)+']<400 && x_dut[0]>-99 && y_dut[0]>-99'
data = root2array(input_file, selection=selection)
```

```python
time_stamp = (data['LP2_10'].T)[ch]
time_ref = (data['gaus_mean'].T)[9]



dt = time_stamp-time_ref

sel = np.logical_and(np.greater(dt,-0.2), np.less(dt, 0.6))
dt = dt[sel]

h = create_TH1D(dt, binning=[100,-1,1])

canvas = rt.TCanvas('c_raw', 'c_raw', 800, 600)
h.Draw()
canvas.Draw()
```

```python
amp = (data['amp'].T)[ch]

risetime = (data['LP2_90'].T)[ch] - (data['LP2_10'].T)[ch]

x_pos = (data['x_dut'].T)[0]
y_pos = (data['y_dut'].T)[0]


x_train = np.column_stack((amp, risetime, x_pos, y_pos))

x_train = x_train[sel]

x_train.shape
```

```python
gbr = GradientBoostingRegressor(loss='ls',
                                n_estimators=500, max_depth=3,
                                learning_rate=.1, min_samples_leaf=4,
                                min_samples_split=3,
                                verbose=1
                                )

gbr.fit(x_train, dt)
```

```python
dt_pred = gbr.predict(x_train)
```

```python
h = create_TH1D(dt - dt_pred, binning=[100,-1,1])

canvas = rt.TCanvas('c_raw', 'c_raw', 800, 600)
h.Draw()
canvas.Draw()
```
